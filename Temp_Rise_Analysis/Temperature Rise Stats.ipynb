# Average Uk Temperature Rise

```python
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


PlottingTemps = []

dataset = Dataset('D:/Data Analysis/tas_hadukgrid_uk_country_ann_188401-201712.nc')
Temps = dataset.variables['tas'][:]
regions = dataset.variables['region'][:]



years = np.arange(1884,2018,1)


for i in range(len(Temps)):
    PlottingTemps.append(Temps[i].mean())

    
    
s = pd.Series(PlottingTemps)
median_val = s.rolling(10).median()
coeffs = np.polyfit(years, PlottingTemps,1)


fitted = coeffs[0]*years + coeffs[1]


fig = plt.figure(figsize=(10,6))      
plt.title('Average UK Temperature Vs Time')
plt.scatter(years,PlottingTemps,label='Original Data')
plt.scatter(years,median_val,label = '10 year rolling mean')
plt.plot(years,fitted,label = ('Linear fit,Gradient='+str(np.round(coeffs[0],2))),linewidth = '5',color= 'green')
plt.legend(loc='upper left')


plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.show()
```

#  Variation in monthly average temp UK

```python
import glob
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

avg= []

dataset = Dataset('D:/Data Analysis/tas_hadukgrid_uk_country_mon_188401-201712.nc')
Temps = dataset.variables['tas'][:]


for item in Temps:
    avg.append(item.mean())

years = np.arange(1884,2018,(1/12))  


fig = plt.figure(figsize = (10,6))
plt.title('Montly Average Temp For UK')
plt.scatter(years,avg)
plt.xlabel('Years')
plt.ylabel('Temperature in (°C)')
plt.show()


s = pd.Series(avg)

rollingstd = s.rolling(120).std()

fig = plt.figure(figsize = (10,6))
plt.title('10 Yr Rolling Standard Deviation ')
plt.xlabel('Years')
plt.ylabel('Standard Deviation')
plt.scatter(years,rollingstd)
plt.show()
    
```
